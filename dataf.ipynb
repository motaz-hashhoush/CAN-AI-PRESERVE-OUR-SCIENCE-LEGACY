```python
import pandas as pd
import os
import glob
import json
```

```python
data_files_names = glob.glob('./Data/*.txt')
meta_files_names = glob.glob('./meta/*.json')
data_files_names = [name[7:]  for name in data_files_names]
meta_files_names = [name[7:]  for name in meta_files_names]


```

```python
new_data = []
new_meta = []
for datafile in data_files_names:
    data_file_id = datafile[:datafile.index('.')]
    for meta_file in meta_files_names:
        meta_file_id = meta_file[:meta_file.index('_')]
        if data_file_id == meta_file_id:
            new_data.append(datafile)
            new_meta.append(meta_file)
            break
    meta_files_names.remove(new_meta[-1])

```

```python
#title , distributionDate, id , authorAffiliations[0].meta.author.name, abstract ,
from operator import contains


dict_list = {"id":[],"title":[],"distributionDate":[],"author":[],"text":[]}
for i in range(len(new_data)):
    dict_data = {}
    with open(f"meta/{new_meta[i]}",'r') as f:
        meta = json.load(f)
    with open(f'Data/{new_data[i]}','r') as f:
        data = f.read()
    dict_list["id"].append(meta['id'])
    dict_list["title"].append(meta['title'])
    dict_list["distributionDate"].append(meta['distributionDate'])
    if 'authorAffiliations' in meta.keys()  :
        dict_list["author"].append(meta['authorAffiliations'][0]['meta']['author']['name'])
    else:
         dict_list["author"].append(None)
    dict_list["text"].append(data)

pd.DataFrame(dict_list).to_csv('data.csv')

    
    
```
